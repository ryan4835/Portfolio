# Drawing a graph to visualise the mixing of signals


```python
import numpy as np 
import networkx as nx
from GCN import GCN
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Parse Karate Club data
train_path = 'C:\\Users\\ryans\Documents/HT/Networks/Datasets/ucidata-zachary/out.ucidata-zachary'
train_file = open(train_path)
train_lines = train_file.readlines()
train_edges = []

for i in train_lines:
    train_edges.append([int(s) for s in i.split() if s.isdigit()][:])

train_edges = train_edges[2:]
train_file.close()
```

```python
G = nx.Graph(train_edges)
test = nx.algorithms.community.centrality.girvan_newman(G)
bipartition = tuple(sorted(c) for c in next(test))
component_1 = bipartition[0]
component_2 = bipartition[1]
cmap = [1 if i in component_1 else 0 for i in list(G)]
graph2 = nx.draw(G, node_color = cmap)
```

```python
network = GCN(G = G, c1 = component_1, c2 = component_2)
network.Mixing_time_average(n_iter = 100)
draw_data = network.Mixing_time_draw()
```
